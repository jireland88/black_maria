# Black Maria Simulations

## Random players

In the first case let's have all players play cards randomly, still following suit, but randomly.

```python
from blackmaria import *
from blackmariaplayers import *

scores = []
for m in range(0,100):
    players = [RandomPlayer(0), RandomPlayer(1), RandomPlayer(2), RandomPlayer(3)]
    table = Table(players)
    table.play_match()
    scores.append(table.get_scores())
```

```python
import numpy as np
import pandas as pd

scores_df = pd.DataFrame.from_records(scores)
print(scores_df.head(5))
```

```python
scores_df.plot(kind="hist", subplots=True)
```

```python
scores_df.boxplot()
```

We can see an average score of just above 150 and an IQR of about 60. Let's create a slightly more intelligent player to see how that fairs.

## The Duck and Dump (Aggressive)

This player will always dump it's highest valued points card (heats / queen of spades) if it has no cards that follow suit (aggressive dumping) or will either play it's highest card following suit below any cards already played, or will just play its lowest card (ducking).

### Simulation 1: 1 DDA, rest random

```python
import sys, os

# Disable print
def blockPrint():
    sys.stdout = open(os.devnull, 'w')

# Restore print
def enablePrint():
    sys.stdout = sys.__stdout__

blockPrint()

scores = []
for m in range(0,100):
    players = [DuckAndDumpAgressive(0), RandomPlayer(1), RandomPlayer(2), RandomPlayer(3)]
    table = Table(players)
    table.play_match()
    scores.append(table.get_scores())
    
scores_df = pd.DataFrame.from_records(scores)
scores_df.boxplot()
```

This new player does fair better against its random opposition, but doesn't crush them. This might be because the DDA player may use up all its low cards early on; ending up stranded with high cards at the end of the round. Before making any new intelligent players let's see what happens if we increase the number of DDA players in the game.
